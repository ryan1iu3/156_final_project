```python
import os
import shutil
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
import time

# Mapping for merging and reducing data set
"""
This mapping show how we consolidated original art styles into five categories to mitigate class imbalance and improve model performance.
"""
reduced_mapping = {
    'Impressionism': 'Impressionism',
    'Post_Impressionism': 'Impressionism',  # Combined with Impressionism
    'Abstract_Expressionism': 'Abstract',
    'Action_Painting': 'Abstract',  # Combined into Abstract
    'Color_Field_Painting': 'Abstract',  # Combined into Abstract
    'Baroque': 'Baroque',
    'Expressionism': 'Expressionism',
    'Realism': 'Realism',
    'New_Realism': 'Realism'  # Combined with Realism
}

reduced_dataset_dir = '/Users/hobelle/Desktop/Math156/reduced_dataset'

# Analyze the distribution of the reduced dataset
"""
Count the number of images in each reduced category
"""
style_counts = {}
for style in os.listdir(reduced_dataset_dir):
    style_path = os.path.join(reduced_dataset_dir, style)
    if os.path.isdir(style_path):  # Ensure it's a directory
        style_counts[style] = len(os.listdir(style_path))  # Count images

style_counts_df = pd.DataFrame(list(style_counts.items()), columns=['Art Style', 'Count'])
print(style_counts_df.sort_values(by='Count', ascending=False))  # Display counts sorted by frequency

# Visualize class distribution
def plot_class_distribution(counts_df, title, color='skyblue'):
    """
    Plots a bar chart for class distribution. The inputs of the fucntion are counts_df which is DataFrame containing art styles and their counts, 
    the title as a string, and the color of the bras as a string. 
    """
    counts_df = counts_df.sort_values(by='Count', ascending=False)  
    plt.figure(figsize=(12, 6))
    plt.bar(counts_df['Art Style'], counts_df['Count'], color=color)
    plt.title(title, fontsize=16)
    plt.xlabel('Art Style', fontsize=12)
    plt.ylabel('Number of Images', fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better readability
    plt.tight_layout()  # Adjust layout for better spacing
    plt.show()
plot_class_distribution(style_counts_df, "Art Styles Distribution (5 Categories)", color='purple')
```
